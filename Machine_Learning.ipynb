# **Machine Learning**
Machine Learning is a field of AI where machines learn patterns from data and make predictions or decisions without explicitly programmed rules.

**Simple meaning:**

*Instead of manually writing rules…*

*We let machine learn from data*

## Traditional Programming:

Rules + Data → Output

## Machine Learning:

Data + Output → Rules learned by Machine

# Features (X)

Features are input variables given to the machine learning model.

These are the values used by the model to make predictions.

# Label / Target (y)

Label (or Target) is the output variable.

This is what the model tries to predict.

**Remember forever:**

*X → Input / Features*

*y → Output / Label / Target*

# **Model**

A model learns patterns from data and makes predictions.


#**Types of Machine Learning**

Supervised Learning

Unsupervised Learning

Reinforcement Learning

# **Supervised Learning**

Supervised Learning is a machine learning approach where model learns from labeled data. (It Maps)

Labeled data means:

Input already has correct output.

Data has inputs + correct outputs. *(Machine learns relation.)*

Used for:

*  Regression
*  Classification

## **Regression**
Output = continuous numeric value

Regression is a machine learning technique used to predict continuous numerical values.

**Continuous values means numbers like:**

*House Price → 50 lakh*

*Salary → 8 lakh*

*Temperature → 32°C*

These values can be any number in a range.

**Goal of Regression**

The goal is to learn relationship between:

Input (X) → Output (y)

Model learns pattern from data and predicts output.

## **Classification**

Classification is a machine learning technique used to predict categories or classes.

*Output is not a number.*

*Output is a label/category.*

**Examples:**

Spam / Not Spam

Yes / No

Fraud / Not Fraud

Cat / Dog

**Goal of Classification**

The goal is to learn relationship between:

Input (X) → Category Output (y)

Model learns patterns and predicts class.

# **Unsupervised Learning**

Unsupervised Learning is a machine learning approach where model learns from unlabeled data. (It find Patterns)

Unlabeled means:

We only have input data.

No correct output is given.

**Used for:**

**Clustering**

Clustering means grouping similar data points together.

Model automatically creates groups based on similarity.


**Pattern Discovery**

Pattern discovery means finding hidden relationships in data.

Model identifies useful patterns automatically.


# **Reinforcement Learning**

Reinforcement Learning is a machine learning approach where an agent learns by trial and error.

Agent takes actions and gets:

Reward → Good action

Penalty → Bad action

Goal:
*Maximize total reward.*

## **Main Components**

1. **Agent**: The learner / decision maker.

**Example:**

Robot

Game AI

Self-driving car

2. **Environment**: The world where agent acts.

**Example:**

Road

Game

Maze

3. **Action**: What agent does.

**Examples:**

Move left

Move right

Stop

4. **Reward**: Feedback after action.

**Example:**

Correct move → +10

Wrong move → -5


# **Training vs Testing**
Extremely important.

Suppose you have 1000 rows.

You don’t train on all data.

**Split into:**

Training Data → Learn

Testing Data → Evaluate

**Example:**

Train = 80%
Test = 20%

# **Overfitting**

Overfitting happens when a model learns training data too well.

It memorizes training data instead of learning general patterns.

As a result:

Performs very well on training data

Performs poorly on testing/new data

# **Machine Learning Pipeline**

Machine Learning Pipeline is a sequence of steps used to build an ML model.

Each step’s output becomes input for next step.

*Step 1 → Step 2 → Step 3 → Final Output*


**Steps in ML Pipeline**

1) **Data Collection**

Collect data from source.

Examples:

CSV files

Database

APIs

2) **Data Preprocessing**

Clean and prepare data.

Tasks:

Handle missing values

Remove duplicates

Encode categorical data

Feature scaling

3) **Feature Selection / Engineering**

Select important features.

Example:

Area, Rooms → Important

House Color → Less important

4) **Train-Test Split**

Split dataset into:

Training Data

Testing Data

Common split:

80% Train

20% Test

5) **Model Training**

Train ML model using training data.

Examples:

Linear Regression

Decision Tree

KNN

6) **Model Evaluation**

Check model performance.

Common metrics:

Accuracy

Precision

Recall

RMSE

7) **Prediction / Deployment**

Use trained model for real-world prediction.

Example:

House Price Prediction

Spam Detection

#Real Dataset Practice

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Age": [25, 30, np.nan, 22, 200, 25],
    "Salary": [50000, np.nan, 60000, 45000, 9999999, 50000],
    "Experience": [2, 4, 3, 1, 100, 2]
}

df = pd.DataFrame(data)
df
```

# Step 1 — Inspect Dataset

```python
#Never clean blindly, Inspect first.
df.head()    #Used to see first 5 rows of dataset.
```

```python
df.shape      #Used to know dataset dimensions.
```

```python
df.info()      #Used to get dataset summary.
```

# Missing Values

```python
df.isnull()    #Check missing values.
```

```python
df.isnull().sum()   #Count missing values.
```

#What to do with missing values?

**Common approaches:**

```python
# 1. Remove rows - Removes rows with missing values.
df.dropna()
```

```python
#2. Fill with value - Fill missing with 0.
df.fillna(0)
```

```python
#3. Fill with mean (Common)
#For Age:
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
```

```python
df.isnull().sum()
```

```python
df
```

**Tips:**

**For numerical columns:**

* mean

* median

**For categorical columns:**

* mode

Common strategy.

# Duplicate Rows


```python
df.duplicated() #Check duplicates.
```

```python
df.duplicated().sum()  #counts dupicate rows
```

```python
df = df.drop_duplicates()  #Remove duplicates
```

```python
df.shape
```

#Invalid Values

```python
#Age = 200 is unrealistic.
df["Age"]
```

```python
df.describe()
```

```python
#We can filter suspicious rows.

df[df["Age"] > 100]
```

```python
#Remove unrealistic ages.
df = df[df["Age"] <= 100]
```

```python
df
```

# Outliers

```python
#Outlier = value far away from normal range.

#Example:
#Salary = 9999999

#Clearly suspicious.
```

```python
df["Salary"]
```

```python
df.describe()
```

```python
df = df[df["Salary"] < 1000000]
```

```python
#Now dataset looks cleaner.
df.describe()
```
