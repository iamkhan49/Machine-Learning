# **Module 4 Machine Learning**
Machine Learning is a field of AI where machines learn patterns from data and make predictions or decisions without explicitly programmed rules.

**Simple meaning:**

## Machine Learning:

# **Supervised Learning**

Supervised Learning is a machine learning approach where model learns from labeled data. (It Maps)

Labeled data means:

Input already has correct output.

Data has inputs + correct outputs. *(Machine learns relation.)*

Used for:

*  Regression
*  Classification

## **Regression**
Output = continuous numeric value

Regression is a machine learning technique used to predict continuous numerical values.

**Continuous values means numbers like:**

*House Price → 50 lakh*

*Salary → 8 lakh*

*Temperature → 32°C*

These values can be any number in a range.

**Goal of Regression**

The goal is to learn relationship between:

Input (X) → Output (y)

Model learns pattern from data and predicts output.

## **Classification**

Classification is a machine learning technique used to predict categories or classes.

*Output is not a number.*

*Output is a label/category.*

**Examples:**

Spam / Not Spam

Yes / No

Fraud / Not Fraud

Cat / Dog

**Goal of Classification**

The goal is to learn relationship between:

Input (X) → Category Output (y)

Model learns patterns and predicts class.

# **Unsupervised Learning**

Unsupervised Learning is a machine learning approach where model learns from unlabeled data. (It find Patterns)

Unlabeled means:

We only have input data.

**Used for:**

**Clustering**

Clustering means grouping similar data points together.

Model automatically creates groups based on similarity.


**Pattern Discovery**

Pattern discovery means finding hidden relationships in data.

Model identifies useful patterns automatically.

Goal:

# **Scaling**

Features with large values dominate.

Scaling: means converting feature values into a similar range.

Scaling = make values comparable.

Two main types
Normalization
Standardization



```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data = {
    "Age": [20, 40, 60]
}

df = pd.DataFrame(data)
df
```

# 1. Normalization (Min-Max Scaling)

**Brings values between:**

0 to 1

**Formula:**

*(x - min) / (max - min)*

**Example:**

**Age values:**

20, 40, 60

**Here:**

min = 20

max = 60

**Convert 20**

*(20-20)/(60-20) = 0/40 = 0*

**Convert 40**

*(40-20)/(60-20) = 20/40 = 0.5*

**Convert 60**

*(60-20)/(60-20) = 40/40 = 1*

**Final:**

20 → 0

40 → 0.5

60 → 1

**Why useful?**

Now values are in same range.

0 to 1

**Best for:**

KNN

K-Means

Because they use distance.

```python
#Apply normalization.

scaler = MinMaxScaler()
df["Age"] = scaler.fit_transform(df[["Age"]])
df
```

# 2. Standardization
*simple : How far is each value from average*

This does NOT force values between 0 and 1.

Negative → below mean

Zero → at mean

Positive → above mean

**Formula:**

*(x - mean) / std*

**Example:**

Age values:

20, 40, 60

**Mean:**

40

**Assume std:**

20

**Convert 20:**

(20-40)/20 = -1

**Convert 40:**

(40-40)/20 = 0

**Convert 60:**

(60-40)/20 = 1

**Final:**

20 → -1
40 → 0
60 → 1

**Interpretation:**

Negative → below mean

Zero → at mean

Positive → above mean

**Best for:**

Linear Regression

Logistic Regression

SVM

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df["Age"] = scaler.fit_transform(df[["Age"]])
df
```

# **Module 12 — Linear Regression**

**Definition:**

Linear Regression is a supervised learning algorithm used to predict continuous numeric values.

**Examples:**

* House Price Prediction

* Salary Prediction

* Sales Prediction

Linear Regression tries to find:

Best-fit straight line through data.

**Example:**

Hours vs Marks

More Hours → More Marks

*Positive relationship.*

# Equation of Linear Regression

**Main equation:**

y = mx + b

Where:

* y = Prediction
* x = Input feature
* m = Slope
* b = Intercept

Example:

Marks = 10 × Hours + 20

If Hours = 5:

Marks = 10×5 + 20 = 70

Prediction = 70

# Meaning of Slope

**Slope tells:**

How much y changes when x increases by 1.

**Example:**

m = 10

**Means:**

Every extra study hour increases marks by 10.

# Meaning of Intercept

**Intercept** = value of y when x = 0

**Example:**

b = 20

**Means:**
At 0 study hours, predicted marks = 20.

# Training in Linear Regression

**Training means:**

**Find best values of:**

* m

* b

Such that line fits data well.

# How do we know line is good?

By checking prediction errors.

**Example:**

**Actual:**

70

**Predicted:**

68

**Error:**

2

Smaller error = better model

All ML algorithms are basically trying to minimize error.

Linear Regression is easiest place to understand this.

```python
# Practical Example
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "Hours": [2, 3, 4, 5],
    "Marks": [40, 50, 60, 70]
}

df = pd.DataFrame(data)
df
```

```python
X = df[["Hours"]]
y = df["Marks"]
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.predict([[6]])
```

```python
model.coef_
```

```python
model.intercept_
```
